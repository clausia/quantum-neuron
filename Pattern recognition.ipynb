```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Data generated in other notebooks

exact = [0.7656249999999982, 0.7656249999999982, 0.5624999999999987,
         0.7656249999999982, 0.015624999999999965, 0.06249999999999986,
         1.3805492484839747e-34, 0.14062499999999967]

q_alg_simple = [0.7655, 0.7613, 0.5645, 0.7681, 0.0171, 0.0613, 0, 0.1402]

q_alg_optimized = [0.7587, 0.7654, 0.56, 0.7669, 0.015, 0.0661, 0, 0.142]
```

```python
# Numbers of sets of bars
sets = len(exact)

# Data on X-axis

# Position of bars on x-axis
ind = np.arange(sets)

plt.figure(figsize=(11,4))

# Width of a bar 
width = 0.2

# Plotting
plt.bar(ind, q_alg_simple , width, label='Simple Algorithm')
plt.bar(ind + width, exact, width, label='Exact Calculation')
plt.bar(ind - width, q_alg_optimized , width, label='Optimized Algorithm')

plt.xlabel('Patterns')
plt.ylabel('Probability')
plt.title('Pattern Recognition')

# xticks()
# First argument - A list of positions at which ticks should be placed
# Second argument -  A list of labels to place at the given locations
plt.xticks(ind + width / 2, ('k=64911', 'k=594', 'k=64957', 'k=610', 'k=16311', 'k=49664', 'k=28415', 'k=30653'))


# Finding the best position for legends and putting it
plt.legend(loc='best')
plt.show()
```
