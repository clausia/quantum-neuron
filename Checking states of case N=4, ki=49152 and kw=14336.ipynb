```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import partial_trace, Statevector
from qiskit.visualization import array_to_latex
from qiskit.circuit.library import ZGate
import numpy as np
```

```python
N = 4
```

## Check circuit for $k_i=49152$

```python
# as seen in Fig 2

circuit_ki = QuantumCircuit(N)

circuit_ki.h(range(N))

circuit_ki.z(0)
circuit_ki.z(1)
circuit_ki.z(2)

circuit_ki.cz(1,2)
circuit_ki.cz(0,2)
circuit_ki.cz(0,1)

threeConrolledZ = ZGate().control(num_ctrl_qubits=2)
circuit_ki.append(threeConrolledZ, [0,1,2])

display(circuit_ki.draw('mpl'))

psi = Statevector(circuit_ki)
display(psi.draw('latex'))

display(array_to_latex(psi.data))
print(psi.data)
```

```python
# only state with phase -1 is |0000>, the other have +1

circuit_0000 = QuantumCircuit(N)

circuit_0000.h(range(N))

circuit_0000.z(0)
circuit_0000.z(1)
circuit_0000.z(2)
circuit_0000.z(3)

circuit_0000.cz(2,3)
circuit_0000.cz(1,3)
circuit_0000.cz(1,2)
circuit_0000.cz(0,3)
circuit_0000.cz(0,2)
circuit_0000.cz(0,1)

threeConrolledZ = ZGate().control(num_ctrl_qubits=2)
circuit_0000.append(threeConrolledZ, [1,2,3])
circuit_0000.append(threeConrolledZ, [0,2,3])
circuit_0000.append(threeConrolledZ, [0,1,3])
circuit_0000.append(threeConrolledZ, [0,1,2])

fourConrolledZ = ZGate().control(num_ctrl_qubits=3)
circuit_0000.append(fourConrolledZ, [0,1,2,3])

display(circuit_0000.draw('mpl'))

psi = Statevector(circuit_0000)
display(psi.draw('latex'))

display(array_to_latex(psi.data))
print(psi.data)
```
